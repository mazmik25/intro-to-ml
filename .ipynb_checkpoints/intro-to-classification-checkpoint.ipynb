````python
# necessary import

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
from sklearn import (datasets,
                   metrics,
                   model_selection as skms,
                   naive_bayes,
                   neighbors)

iris = datasets.load_iris()
(iris_train_ftrs, iris_test_ftrs,
 iris_train_trgts, iris_test_trgts) = skms.train_test_split(iris.data,
                                                               iris.target,
                                                               test_size=.95)
len(iris_train_ftrs), len(iris_test_ftrs)
````

````python
iris_df = pd.DataFrame(iris.data,
                       columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df.head(3)
````

````python
iris_df.describe()
````

````python
iris_df.target.value_counts()
````

````python
sns.pairplot(iris_df, hue='target', height=1.5)
````

# Classification

## Naive Bayes and kNN for Classification

### Naive Bayes

Where:
- **P(A|B)** is the probability of hypothesis A given the data B. This is called the [_posterior probability_](https://www.statisticshowto.com/posterior-distribution-probability/).
- **P(B|A)** is the probability of data B given that the hypothesis A was true.
- **P(A)** is the probability of hypothesis A being true (regardless of the data). This is called the prior probability of A.
- **P(B)** is the probability of the data (regardless of the hypothesis).

````python
prefs = """cat tea red
            cat coffee red
            cat tea red
            dog coffee red
            cat tea black
            dog coffee black
            dog coffee black
            dog tea black""".split()
df = pd.DataFrame(np.array(prefs).reshape(8,3),
                  columns=["pet", "drink", "color"])
df
````

````python
pd.crosstab(df.pet, df.color)
````

````python
red_data = df.groupby('color').get_group('red')

red_data.pet.value_counts(normalize=True)
````

````python
red_data.drink.value_counts(normalize=True)
````

````python
df.color.value_counts(normalize=True)
````

4. ======> compare the value with the black probabilites
5. result => it might be red

````python
nb = naive_bayes.GaussianNB()
fit = nb.fit(iris_train_ftrs, iris_train_trgts)
preds = fit.predict(iris_test_ftrs)

actuals = iris_test_trgts
metrics.accuracy_score(actuals, preds)
````

### KNN for Classification


kNN can be interpreted in the real life. People tend to be effected by the people around them. Our behaviour is guided by the friends we grew up with. Our parents also shape our personality in some ways. If you grow up with people who love sports, it is higly likely that you will end up loving sports. There are -- of course -- exceptions.

```
The value of a data point is determined by the data points around it.
```

````python
ftrs = np.random.randint(1, 7, (8,2))
ftrs
````

````python
trgts = np.random.randint(0,2,8)
trgts
````

````python
plt.scatter(ftrs[:,0], ftrs[:,1], c=trgts)
tst = (2,3)
plt.scatter(*tst, c='r')
````

````python
# c^2 = a^2 + b^2
# a: difference in xs between me and someone else
# b: difference in ys between me and someone else
distances = np.hypot(2-ftrs[:,0], 3-ftrs[:,1])
distances
````

````python
distances.argsort()
````

````python
iris.target_names
````

````python
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
fit = knn.fit(iris_train_ftrs, iris_train_trgts)
preds = fit.predict(iris_test_ftrs)

actuals = iris_test_trgts
metrics.accuracy_score(actuals, preds)
````
